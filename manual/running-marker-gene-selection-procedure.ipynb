### Running marker gene selection procedure

Expected input:
1. ST gene expression matrix.
2. Candidate marker genes for each expected cell type in `.txt` files. Each cell type should have its own ".txt" file in a folder named `candidate-markers` in `input_data ` directory.

### Loading exemplificatory ST data 

```r
library(Seurat)
library(SeuratData)
#InstallData("stxBrain")

brain <- LoadData("stxBrain", type = "anterior1")
ST <- brain@assays$Spatial@data

#ST <- ST[rownames(ST) %in% rownames(B), ]
ST <- as.matrix(ST)
ST[1:5,1:5]
```

### Files wth candidate marker genes

```r
list.files(paste0(input_data, "candidate-markers"))
```

```r
head(read.table(paste0(input_data, "candidate-markers", "/", "ASC.txt")))
```

For each gene we choose the *core gene* and we set $\tau$ and $\rho$ parameters.

```r
cell_types <- c("ASC", "CPC", "GABA", "GLUT", "OLG")
lead_genes <- c("Gja1", "Ttr", "Ppp1r1b", "Cck","Plp1")
tresh_tau <- c(0.5,0.7,0.7,0.7, 0.7)
tresh_rho <- c(0.5,0.7,0.7,0.7, 0.7)



# source function marker-gene-selection from code directory in Celloscope repository
source("~./Celloscope/code/marker-gene-selection.R")
address_candidates <- paste0(input_data, "/candidate-markers/")
matB <- prepare_matrix_B(ST, address_candidates, cell_types, lead_genes, tresh_tau, tresh_rho )
#adding dummy type
matB <- cbind(matB, 0)
cell_types_with_DT <- c(cell_types, "DT")
colnames(matB) <- cell_types_with_DT
dir.create(paste0(input_data, "/gene-selection-output"), showWarnings = FALSE)
### matB.csv should appear in gene-selection-output directory
write.csv(matB, paste0(input_data, "/gene-selection-output/matB.csv"))
```
